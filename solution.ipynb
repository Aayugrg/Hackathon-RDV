```python
#importing necessary libraries
import pandas as pd
import tensorflow as tf
from tensorflow.keras.layers import *
from tensorflow.keras.models import *
import numpy as np
from sklearn.preprocessing import LabelEncoder


```

```python
#getting the required dataset
dataset=pd.read_csv(r"C:\Users\HP PC\Downloads\data.csv")
dataset
```

```python
# First Model (Predicting species Based on fingers and tails)
```

```python
#extracting the message column from dataset and storing in X_messages
X_messages=dataset["message"]
X_messages
```

```python
#creating output label have species column
Y_species=dataset["species"]
Y_species
```

```python
#creating input label for the first part of the model (fingers+tail)
X=dataset.drop(["message","species"],axis=1)
X
```

```python
#one-hot-encoding tail variable and converting into integer for model
X = pd.get_dummies(X, columns=["tail"])
X=X.astype(int)
X
```

```python
#labelling 10 species from 0-10 to use softmax

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform the Y labels automatically
Y = label_encoder.fit_transform(Y_species)
species_mapping = dict (zip(label_encoder.classes_, range(len(label_encoder.classes_))))

# Print the encoded labels and corresponding classes
print("Encoded Labels: ", Y)
print("Classes: ", label_encoder.classes_)  # Shows the mapping
print(species_mapping)
```

```python
# Define a deep neural network for species prediction based on fingers and tail data
model = tf.keras.Sequential([
    Dense(1024, activation='relu', input_shape=(3,), kernel_regularizer=tf.keras.regularizers.l2(0.001)),
    BatchNormalization(),
   Dense(512, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001)),
     BatchNormalization(),
    Dense(256, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001)),
     BatchNormalization(),
    Dense(128, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001)),
     BatchNormalization(),
    Dense(64, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001)),
     BatchNormalization(),
    Dense(32,activation="relu",kernel_regularizer=tf.keras.regularizers.l2(0.001)),
     BatchNormalization(),
    Dense(10,activation="linear")
])

# Compile the model using the Adam optimizer and sparse categorical cross-entropy loss
# Sparse categorical cross-entropy is used since Y is label encoded (integer labels)
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

model.summary()
```

```python
# Train the model using the input data X (fingers + tail) and output labels Y (species)
# We use 20% of the data for validation, with 500 epochs and a batch size of 100
history = model.fit(X, Y, epochs=500, batch_size=100, validation_split=0.2)
```

```python
# Predicting the species based on the input data (fingers + tail)
answer=model.predict(X)
answer
```

```python
# Convert the output logits to probabilities using softmax
final_answer=tf.nn.softmax(answer)
final_answer
```

```python
# Initialize a dictionary to store word-level probabilities for each species based on message content
df = np.array(dataset)
dic = {}
for i in df:
    msg = i[0].split()
    for word in msg:
        if word not in dic:
            dic[word]={i[3]:1}
        else:
            dic[word][i[3]] = dic[word].get(i[3],0) + 1
# Normalize the probabilities for each word by dividing by total occurrences of the word
for word in dic:
    occurences = sum([dic[word][i] for i in dic[word]])
    for species in dic[word]:
        dic[word][species]/=occurences

print(dic)
```

```python
# Extract words that are 100% associated with a single species (sure match)
sure_dict={}
for key, sub_dict in dic.items():
    # Check if any probability is exactly 1.0
    for sub_key, prob in sub_dict.items():
        if prob == 1.0:
            sure_dict[key] = sub_key
            break

sure_dict

```

```python
# Helper function to get message-based probabilities for species prediction
def get_message_probabilities(message, dic,species_mapping):
    words = message.split()
    species_prob = np.zeros(10)  # Initialize an array for 10 species (size 10 because you have 10 species)

    for word in words:
        if word in dic:
            for species_label in dic[word]:
                species_index=species_mapping[species_label]


                species_prob[species_index] += dic[word][species_label]  # Add probabilities for each species



    return species_prob/np.sum(species_prob)


# Helper function to get finger/tail-based probabilities
def get_finger_tail_probabilities(X_row, dict):
    X_tuple = tuple(X_row)
    if X_tuple in dict:
        return dict[X_tuple]
    else:
        return np.ones(10) / 10  # Return uniform probability if no match

# Combine message and finger/tail probabilities using weights alpha and beta
def combined_probability(message, X_row, dic, dicti, alpha=0.9, beta=0.1):
    message_prob = get_message_probabilities(message, dic,species_mapping)
    finger_tail_prob = get_finger_tail_probabilities(X_row, dicti)

    final_prob = np.power(message_prob,alpha) + np.power(finger_tail_prob,beta)
    return final_prob / np.sum(final_prob)  # Normalize to get probabilities

```

```python
# Example input
message = "cosmix xeno nebuz odbitaz"  # Test message
fingers = 5  # Test fingers value
X_row=[5,0,1]


# Fetch probabilities from the combined model
def predict_species_test(message, X_row, dic, dicti, species_mapping):
    # Get probabilities from the message
    message_prob = get_message_probabilities(message, dic,species_mapping )

    # Get probabilities from finger/tail data
    finger_tail_prob = get_finger_tail_probabilities(X_row, dicti)

    # Use the combined model to compute the final probabilities
    #alpha beta found out after manual hit and trial
    final_prob = combined_probability(message, X_row, dic, dicti, alpha=0.92, beta=0.08)

    # Find the species with the highest probability (argmax)
    predicted_species_index = np.argmax(final_prob)

    # Return the predicted species label using label_encoder
    return predicted_species_index

# Test the model with the input provided
predicted_species_index = predict_species_test(message, X_row, dic, dicti,species_mapping)
predicted_species_index

```
